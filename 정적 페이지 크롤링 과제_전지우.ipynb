#### 1. cgv 영화 예매 순위 크롤링

```python
import urllib.request as request
from bs4 import BeautifulSoup
url = request.urlopen("http://www.cgv.co.kr/movies/")
soup = BeautifulSoup(url,"html.parser")
```

```python
def movie_information(movie):
    rank = movie.find("div",{"class":"box-image"})
    rank_strong = rank.find("strong",{"class":"rank"}).text[3]
    
    contents = movie.find("div",{"class":"box-contents"})
    tag = contents.find("a")
    name = tag.find("strong").text
    
    score = movie.find("div",{"class":"score"})
    strong = score.find("strong")
    score_rate = strong.find("span").text
    
    return {'순위':rank_strong+'등','이름':name,'예매율':score_rate}
```

```python
def box(cgv):
    list = cgv.find_all("li")
    movie_info = [movie_information(movie) for movie in list]
    return movie_info
```

```python
chart = soup.find("div",{"class":"sect-movie-chart"})
cgv = chart.find_all("ol")
```

```python
cgv
```

```python
for i in range(2):
    movie_info = box(cgv[i])
    for info in movie_info:
        print(info)
```

#### 2. 넷플릭스 영화 줄거리 크롤링(200개만 크롤링)

```python
from bs4 import BeautifulSoup
import urllib.request
from time import sleep
from random import randint
import requests
import pandas as pd
import lxml.html
import numpy as np
import urllib
import re
from time import sleep
from random import randint
```

```python
df = pd.DataFrame(columns=["title","genre","netflix"])
count = 0
```

```python
for num in range(2,200):
    url = "https://www.4flix.co.kr/board/netflix/" + str(num)
    sleep(randint(1,5))
    with urllib.request.urlopen(url) as url:
        try:
            doc = url.read()
            soup = BeautifulSoup(doc, "html.parser")
        
            title_year = soup.find_all("h1")[2].text.strip() # 제목(연도) 
            title = title_year[:-6] #제목만
            genre = soup.find_all("h3")[1].text.strip()
            netflix = soup.find_all("p")[0].text.strip()
        
            df.loc[count] = [title, genre, netflix]
            count+=1
        except:
            pass
```
