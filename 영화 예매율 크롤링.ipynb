```python
def movie_information(movie):
    rank = movie.find("div",{"class":"box-image"})
    rank_strong = rank.find("strong",{"class":"rank"}).text[3]
    rank.append(rank_strong)
    
    contents = movie.find("div",{"class":"box-contents"})
    tag = contents.find("a")
    name = tag.find("strong").text
    title.append(name)
    
    score = movie.find("div",{"class":"score"})
    strong = score.find("strong")
    score_rate = strong.find("span").text
    rate.append(score_rate)
    
    return
```

```python
def box(cgv):
    list = cgv.find_all("li")
    movie_info = [movie_information(movie) for movie in list]
    return movie_info
```
